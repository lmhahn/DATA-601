```python
#Create 10 random Credit-Scores between 100 and 800. 
creditscores = []
for i in range(0,10):
    x = random.randint(100, 800)
    creditscores.append(x)
print(creditscores)
```

```python
#Using nested loop comprehension to generate Credit Approval list
#Using range function to perform an iteration for each object in the "creditscores" list
#First choosing to iterate through list of credit scores and assign 'no' to any values <= 650
#Using nested conditional statements to then interate through list of birth years to determine age eligibility
#Using append function to add each 'answer' to the credit approval list

creditapproval = []

for i in range(len(creditscores)):
    for score in creditscores:
        if score <= 650:
            answer = 'no'
        else:
            for year in byears:
                if year >= 2007:
                    answer = 'no'
                else:
                    answer = 'yes'
    creditapproval.append(answer)

creditapproval
```
